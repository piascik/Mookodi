```python
import numpy as np
from astropy.io import fits
```

```python
#print(nrand)
#print(nbias)
#print(ndark)
#print(nflat)
```

```python
# Random noise, 100 ± 10 
nrand = np.random.normal(100, 10, size=(1024, 1024))       
hdu = fits.PrimaryHDU(nrand)
hdu.header['EXPOSURE'] = 10
hdu.writeto('randomimage.fits',overwrite='True')
```

```python
# normal 10 ± 1
nbias = np.random.normal(0, 1, size=(1024, 1024)) + 10     
hdu = fits.PrimaryHDU(nbias)
hdu.writeto('zeroimage.fits',overwrite='True')
```

```python
# random between 0 and 10. Never -ve
ndark = np.random.uniform(0, 10, size=(1024, 1024))         
hdu = fits.PrimaryHDU(ndark)
hdu.header['EXPOSURE'] = 100
hdu.writeto('darkimage.fits',overwrite='True')
```

```python
# normal 1 ± 0.01
nflat = np.random.normal(0, 0.01, size=(1024, 1024)) + 1   
hdu = fits.PrimaryHDU(nflat)
hdu.writeto('flatimage.fits',overwrite='True')
```

```python
import matplotlib.pyplot as plt
mean = [0, 0]
cov = [[1, 0], [0, 1]]
x, y = np.random.multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```
